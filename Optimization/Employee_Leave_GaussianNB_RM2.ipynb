```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# pulling the csv into a dataframe
employee_df = pd.read_csv('./resources/Employee.csv')
# displaying the data drame
employee_df.head()
```

```python
# converting the columns with words into boolen
employee_city = pd.get_dummies(employee_df[['Education','City', 'Age', 'EverBenched']])
#displaying the dataframe
employee_city
```

```python
# joining the Two dataframes 
Joined_df = pd.merge(employee_df,employee_city,)
Joined_df.head()
```

```python
# Seperate the features, X,  from the target variable, y


y = np.array(Joined_df[['LeaveOrNot']])
X = np.array(Joined_df.drop(columns=['Education','City', 'Age', 'EverBenched','Gender','LeaveOrNot']))
```

```python
# testing the data

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# scaling the data
scaler = StandardScaler()
X_scaler = scaler.fit(X_train)
X_train = X_scaler.fit_transform(X_train)
X_test = X_scaler.transform(X_test)
```

```python
# the creating model code was pulled from the following URL https://builtin.com/artificial-intelligence/gaussian-naive-bayes
```

```python
#creating  the model 
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.score(X_test, y_test)
```

```python
Y_var_pred= model.predict(X_test)
```

```python
print(classification_report(y_test, Y_var_pred))
```
